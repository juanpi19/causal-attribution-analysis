1. Model a causal inference problem using assumptions.
2. Identify an expression for the causal effect under these assumptions (“causal estimand”)
3. Estimate the expression using statistical methods such as matching or instrumental variables
4. Finally, verify the validity of the estimate using a variety of robustness checks

```python
# Create a new directed graph
G = nx.DiGraph()


# Add nodes
nodes = [
    "origin", "first_contact_date", "is_qualified", "won_date", "seller_state",
    "seller_city", "product_category_name", "product_size",
    "order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date",
    "order_delivered_customer_date", "order_estimated_delivery_date", "order_status",
    "review_score", "payment_value", "product_category_name", "price", "freight_value",
    "distance_seller_customer", "days_in_delivery", "is_late_delivery", "is_product_right_U"

]
G.add_nodes_from(nodes)


# Add edges (relationships)
edges = [
    ("origin", "first_contact_date"),
    ("first_contact_date", "is_qualified"),
    ("is_qualified", "won_date"),
    ("won_date", "seller_state"),
    ("won_date", "seller_city"),
    ("seller_state", "seller_city"),
    ("product_category_name", "product_size"),
    ("order_purchase_timestamp", "order_approved_at"),
    ("order_approved_at", "order_delivered_carrier_date"),
    ("order_delivered_carrier_date", "order_delivered_customer_date"),
    ("order_delivered_carrier_date", "order_estimated_delivery_date"),
    ("order_delivered_customer_date", "order_status"),
    ("order_delivered_customer_date", "is_late_delivery"),
    ("is_late_delivery", "review_score"),
    ("order_status", "review_score"),
    ("days_in_delivery", "review_score"),
    ("is_product_right_U", "review_score"),
    ("product_category_name", "payment_value"),
    ("product_category_name", "price"),
    ("product_category_name", "freight_value"),
    ("distance_seller_customer", "days_in_delivery")


]
G.add_edges_from(edges)

# Set up the plot
plt.figure(figsize=(18, 10))
pos = nx.spring_layout(G, k=0.9, iterations=50)

# Draw the graph
nx.draw(G, pos, with_labels=True, node_color='lightblue', 
        node_size=3000, font_size=8, font_weight='bold', 
        arrows=True, edge_color='gray')

# Add edge labels
edge_labels = {(u, v): '' for (u, v) in G.edges()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=6)

# Show the plot
plt.title("Olist Causal DAG", fontsize=16)
plt.axis('off')
plt.show()

```

```python
def haversine_distance_vectorized(lat1, lon1, lat2, lon2):
    """takes two sets of tuples and calculates haversine distance to return distance in km"""
    R = 6371  # Earth's radius in kilometers

    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    distance = R * c

    return distance


def apply_haversine_distance(df, lat1, lon1, lat2, lon2):
    df['distance_in_km'] = haversine_distance_vectorized(df[lat1], df[lon1], df[lat2], df[lon2])
    return df

```

```python
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])

df['late_delivery_in_days'] = (df['order_delivered_customer_date'] \
                                            - df['order_estimated_delivery_date']).dt.days

# if late_delivery_in_days > 0, then delivery was late
df['is_delivery_late'] = np.where(df['late_delivery_in_days'] > 0, 1, 0)
```

```python
olist_geolocation_df.drop_duplicates(subset='geolocation_zip_code_prefix', inplace=True)

# merging to get latitude and longitude for seller
df = df.merge(
    olist_geolocation_df[['geolocation_zip_code_prefix', 'geolocation_lat', 'geolocation_lng']]\
        .rename(columns={'geolocation_lat': 'seller_geolocation_lat', 'geolocation_lng': 'seller_geolocation_lng'}),
          left_on='seller_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left'
        )

# merging to get latitude and longitude for customer
df = df.merge(
    olist_geolocation_df[['geolocation_zip_code_prefix', 'geolocation_lat', 'geolocation_lng']]\
        .rename(columns={'geolocation_lat': 'customer_geolocation_lat', 'geolocation_lng': 'customer_geolocation_lng'}),
          left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left'
        )
```

```python
df = apply_haversine_distance(df, 'seller_geolocation_lat', 'seller_geolocation_lng', 'customer_geolocation_lat', 'customer_geolocation_lng')
```

```python
columns = ['product_category_name', 'freight_value', 'is_delivery_late', 'distance_in_km', 'review_score']
temp_df = df[columns]
temp_df.head()
```

```python
temp_df.dropna(inplace=True)
```

```python
causal_graph = """digraph {
is_delivery_late;
freight_value;
product_category_name;
distance_in_km;
review_score;

freight_value->is_delivery_late;
freight_value -> review_score;

is_delivery_late -> review_score;

distance_in_km->is_delivery_late;
distance_in_km -> review_score;

product_category_name->is_delivery_late;
product_category_name -> review_score;
}"""
```

```python
model = CausalModel(data=temp_df,
                          graph=causal_graph.replace("\n", " "),
                          treatment="is_delivery_late",
                          outcome="review_score")
model.view_model()
```

```python
identified_estimand = model.identify_effect(proceed_when_unidentifiable=True)
print(identified_estimand)
```

```python
causal_estimate_ipw = model.estimate_effect(identified_estimand,
                                            method_name="backdoor.propensity_score_weighting",
                                            target_units = "ate")

print(causal_estimate_ipw)
print("Causal Estimate is " + str(causal_estimate_ipw.value))
```

On average, when a delivery is late (is_delivery_late = 1), the review score decreases by approximately 1.88 points, holding other factors constant.

```python
refutation = model.refute_estimate(identified_estimand, 
                                   causal_estimate_ipw, 
                                   method_name='placebo_treatment_refuter',
                                   placebo_type='permute', 
                                   num_simulations=20)
print(refutation)
```

**For a placebo test in causal inference:**

Null Hypothesis (H0):

The null hypothesis in a placebo test is that there is no real causal effect of the treatment on the outcome. In other words, the observed effect in your original analysis is due to chance or model misspecification, not a true causal relationship.
Formally, H0: The distribution of estimated effects under the placebo treatment is the same as the distribution that generated your observed treatment effect.

Alternative Hypothesis (H1):

The alternative hypothesis is that there is a real causal effect of the treatment on the outcome, which is different from what would be observed under random (placebo) assignment of the treatment.
Formally, H1: The distribution of estimated effects under the placebo treatment is different from the distribution that generated your observed treatment effect.

# Bootstrap simulation for Power Analysis

https://www.theeffectbook.net/ch-Simulation.html

```python
# Boostrap Simulation for Power Analysis

alpha = 0.05
sample_sizes = [i for i in range(1,1000,30)]
results = {}

for sample in sample_sizes:
    results[sample] = []
    for sim in range(100):
        indexes = np.random.choice(temp_df.index, size=sample)
        bootstrap_df = temp_df.loc[indexes]
        ols_model = smf.ols('review_score ~ is_delivery_late + product_category_name + distance_in_km + freight_value', data=bootstrap_df).fit()
        is_significant = ols_model.pvalues['is_delivery_late'] < 0.05
        results[sample].append(is_significant)
    print(sample)
```

```python
power = [sum(results[i])/100 for i in sample_sizes]

plt.plot(sample_sizes, power)
plt.title("Power Curve")
plt.xlabel("sample size")
plt.ylabel("power")
plt.show()
```
